```python
import numpy as np
import mne

# Load EDF file into memory
edf = mne.io.read_raw_edf('Subject00_1.edf', preload=True)

# Get total duration in seconds
duration = edf.times[-1]

# Find midpoint
midpoint = duration / 2.0

# Define 60-second window around the middle
tmin = max(0, midpoint - 30)
tmax = min(duration, midpoint + 30)

# Crop to middle 60 seconds
edf.crop(tmin=tmin, tmax=tmax)

# Get data (channels x samples)
data = edf.get_data().T

# Build time vector (length = number of samples, step = 1/sampling rate)
times = edf.times[:data.shape[0]].reshape(-1, 1)

# Combine time column with data
data_with_time = np.hstack((times, data))

# Build header: Time + channel names
header = 'Time,' + ','.join(edf.ch_names)

# Save to CSV
np.savetxt('Subject00_1_middle60s.csv', data_with_time, delimiter=',', header=header, comments='')
```
